```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df= pd.read_csv('mtcars.csv')
df
```

```python
hist=sns.histplot(x='mpg',data=df,bins=10,kde=True)
hist
```

```python
Scatter= sns.scatterplot(x='wt',y='mpg',hue='mpg',data=df)
Scatter
```

```python
bar=sns.countplot(x='am',data=df)
bar
```

```python
box=sns.boxplot(x='mpg',data=df)
box
```
